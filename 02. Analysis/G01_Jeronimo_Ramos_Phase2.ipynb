# Exercise 2: Supervised learning.

```python
# First we will import the libraries we will use
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

## 1. Models used for regression

The first model we will be making is the linear regression. This model allows us to determine a line that better fits to the input data and it's able to correctly predict an output value according to a new data.

Then, we will make a regression model with the Ridge (L2) and Lasso (L1) regularizations to see the behaviour of the models by comparing the quality of the models using the Mean Squared Error (MSE), Mean Absolute Error (MAE), $R^2$ score and Root Mean Squared Error (RMSE).

However, we will explore the behavior of the Ridge and Lasso regularizations in order to obtain the best value of the hyper paramether: $\alpha$. This is important because the value of $\alpha$ determines the force of the regulation. If $\alpha$ is near to zero, the model will act as a Linear Regression, but if $\alpha$ is high, the coefficients will be drastically reduced, which can lead to an underfitting of the model.

### Data splitting

To begin with, we will split the data into two subsets of data to train the model and to test it afterwards and see if the model can generalize correctly the data. For this, we define the loss function as the Mean Squared Error (MSE)

```python
X = df_reg['data']
y = df_reg['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

### Calculating the value of $\alpha$

```python
# First we define a set of values for Alpha in a logaritmic scale to see the changes
alphas = [0.001, 0.01, 0.1, 1, 10, 100]

ridge_r2 = []
lasso_r2 = []

for a in alphas:
    # Pipeline for Ridge
    ridge_pipe = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=a))])
    ridge_pipe.fit(X_train, y_train)
    ridge_r2.append(ridge_pipe.score(X_test, y_test))
    
    # Pipeline for Lasso
    lasso_pipe = Pipeline([('scaler', StandardScaler()), ('model', Lasso(alpha=a))])
    lasso_pipe.fit(X_train, y_train)
    lasso_r2.append(lasso_pipe.score(X_test, y_test))

#Note: We scale the data so the difference in the magnitude of the data doesn't affect the Lasso or the ridge regularization

# Then we present the behavior of the R2 metric according to the different values of alpha
plt.figure(figsize=(10, 6))
plt.plot(alphas, ridge_r2, marker='o', label='Ridge (R²)', color='blue')
plt.plot(alphas, lasso_r2, marker='s', label='Lasso (R²)', color='red')

plt.xscale('log') # Escala logarítmica para mejor visualización
plt.xlabel('Alpha value (logarithmic scale)')
plt.ylabel('R² Score (Precision)')
plt.title('Alpha\'s Effect in the Model Precision')
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.show()
```

As we can see in the previous graph, the ridge regularization tends to be more stable because the coefficients do not reach zero. However, Lasso tends to lose accuracy faster as Alpha increases because it starts to "turn off" variables. This means that the best Alpha is the one where the coefficient is at its maximum in the test set.

In order to know the best value of alpha for both cases, we will use the tool `GridSearchCV` which is a class from the scikit-learn library used for hyperparameter tuning. It systematically looks out for the optimal combination of hyperparameter values given a machine learning model and a list of possible values for each hyperparameter by using cross-validation to evaluate the performance of each combination.

```python
# First we define the possible values of alpha or the grid for the class
param_grid = {'model__alpha': np.logspace(-4, 4, 20)}

# Then we configure the search for the Ridge Regression 
ridge_search = GridSearchCV(
    Pipeline([('scaler', StandardScaler()), ('model', Ridge())]),
    param_grid, cv=5, scoring='r2'
)

# And the Lasso regression
lasso_search = GridSearchCV(
    Pipeline([('scaler', StandardScaler()), ('model', Lasso(max_iter=10000))]),
    param_grid, cv=5, scoring='r2'
)

# Note: In both cases we are using a 5-fold Cross-Validation and we are using the R2 metric to
#  evaluate the performance of the cross-validated model on the test set

# Now we perform the shearch
ridge_search.fit(X_train, y_train)
lasso_search.fit(X_train, y_train)

# Results
print(f"Best alpha for Ridge: {ridge_search.best_params_['model__alpha']:.4f}")
print(f"Best R2 in validation (Ridge): {ridge_search.best_score_:.4f}")
print("-" * 30)
print(f"Best alpha for Ridge: {lasso_search.best_params_['model__alpha']:.4f}")
print(f"Best R2 in validation (Lasso): {lasso_search.best_score_:.4f}")
```

### Comparing the models

Now that we know the best values for $\alpha$ we will use them to compare the metrics of each model

```python
models = {"Linear Regression": LinearRegression(),
          "Lasso": Lasso(alpha=0.0886),
          "Ridge": Ridge(alpha=1.6238)
          }

results = []

for name, model in models.items():
    pipe = Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    # Calculate the metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    results.append([name, mse, mae, rmse, r2])

    

# We create the comparative table
df_results = pd.DataFrame(results, columns=['Model', 'MSE', 'MAE', 'RMSE', 'R2'])
print(df_results.to_markdown(index=False))
```
