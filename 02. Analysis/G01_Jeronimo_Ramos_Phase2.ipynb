# Exercise 2: Supervised learning.

```python
# First we will import the libraries we will use
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

## Data splitting
To begin with, we will split the data into two subsets of data to train the model and to test it afterwards and see if the model can generalize correctly the data. For this, we define the loss function as the Mean Squared Error (MSE)

```python
X = df_reg['data']
y = df_reg['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## 1. Models used for regression
The first model we will be making is the linear regression. This model allows us to determine a line that better fits to the input data and it's able to correctly predict an output value according to a new data.

Then, we will make a regression model with the Ridge (L2) and Lasso (L1) regularizations to see the behaviour of the models by comparing the quality of the models using the Mean Squared Error (MSE), Mean Absolute Error (MAE), R^2 score and Root Mean Squared Error (RMSE).

```python
# Definir rango de Alphas (escala logarítmica para ver cambios sutiles y drásticos)
alphas = [0.001, 0.01, 0.1, 1, 10, 100]

ridge_r2 = []
lasso_r2 = []

for a in alphas:
    # Pipeline para Ridge
    ridge_pipe = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=a))])
    ridge_pipe.fit(X_train, y_train)
    ridge_r2.append(ridge_pipe.score(X_test, y_test))
    
    # Pipeline para Lasso
    lasso_pipe = Pipeline([('scaler', StandardScaler()), ('model', Lasso(alpha=a))])
    lasso_pipe.fit(X_train, y_train)
    lasso_r2.append(lasso_pipe.score(X_test, y_test))

# Graficar el comportamiento de R2 según Alpha
plt.figure(figsize=(10, 6))
plt.plot(alphas, ridge_r2, marker='o', label='Ridge (R²)', color='blue')
plt.plot(alphas, lasso_r2, marker='s', label='Lasso (R²)', color='red')

plt.xscale('log') # Escala logarítmica para mejor visualización
plt.xlabel('Valor de Alpha (escala logarítmica)')
plt.ylabel('Puntaje R² (Precisión)')
plt.title('Efecto de Alpha en la Precisión del Modelo')
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.show()
```

```python
models = {"Linear Regression": LinearRegression(),
          "Lasso": Lasso(alpha=0.0886),
          "Ridge": Ridge(alpha=1.6238)
          }

results = []

for name, model in models.items():
    pipe = Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    # Calculate the metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    results.append([name, mse, mae, rmse, r2])

# We create the comparative table
df_results = pd.DataFrame(results, columns=['Model', 'MSE', 'MAE', 'RMSE', 'R2'])
print(df_results.to_markdown(index=False))
```

```python
# 1. Graficar la importancia de los coeficientes (usando el modelo Ridge del ejercicio anterior)
# Extraemos los nombres de las columnas y sus respectivos coeficientes
feature_names = X.columns
model_coefficients = pipe.named_steps['model'].coef_

plt.figure(figsize=(12, 5))

# Subplot 1: Coeficientes
plt.subplot(1, 2, 1)
coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': model_coefficients})
coef_df = coef_df.sort_values(by='Coefficient', ascending=False)
sns.barplot(x='Coefficient', y='Feature', data=coef_df, palette='viridis')
plt.title('Importancia de los Coeficientes (Modelo Ridge)')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Subplot 2: Valores Reales vs Predicciones
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred, alpha=0.6, color='blue', edgecolors='k')
# Línea de regresión ideal (donde predicción == real)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reales (BodyFat %)')
plt.ylabel('Predicciones (BodyFat %)')
plt.title('Ajuste del Modelo: Real vs Predicho')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# 1. Definir el rango de valores para Alpha (desde muy pequeños hasta grandes)
param_grid = {'model__alpha': np.logspace(-4, 4, 20)}

# 2. Configurar la búsqueda para Ridge
ridge_search = GridSearchCV(
    Pipeline([('scaler', StandardScaler()), ('model', Ridge())]),
    param_grid, cv=5, scoring='r2'
)

# 3. Configurar la búsqueda para Lasso
lasso_search = GridSearchCV(
    Pipeline([('scaler', StandardScaler()), ('model', Lasso(max_iter=10000))]),
    param_grid, cv=5, scoring='r2'
)

# Ejecutar la búsqueda
ridge_search.fit(X_train, y_train)
lasso_search.fit(X_train, y_train)

# Resultados
print(f"Mejor Alpha para Ridge: {ridge_search.best_params_['model__alpha']:.4f}")
print(f"Mejor R2 en validación (Ridge): {ridge_search.best_score_:.4f}")
print("-" * 30)
print(f"Mejor Alpha para Lasso: {lasso_search.best_params_['model__alpha']:.4f}")
print(f"Mejor R2 en validación (Lasso): {lasso_search.best_score_:.4f}")
```
