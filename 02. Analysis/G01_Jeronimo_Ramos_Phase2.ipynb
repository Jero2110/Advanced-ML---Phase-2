## Models used for regression

The first model we will be making is the linear regression. This model allows us to determine a line that better fits to the input data and it's able to correctly predict an output value according to a new data.

Then, we will make a regression model with the Ridge (L2) and Lasso (L1) regressions to see the behaviour of the models by comparing the quality of the models using the Mean Squared Error (MSE), Mean Absolute Error (MAE), $R^2$ score and Root Mean Squared Error (RMSE).

However, we will explore the behavior of the Ridge and Lasso regressions in order to obtain the best value of the hyper paramether: $\alpha$. This is important because the value of $\alpha$ determines the force of the regulation. If $\alpha$ is near to zero, the model will act as a Linear Regression, but if $\alpha$ is high, the coefficients will be drastically reduced, which can lead to an underfitting of the model.

```python
# First we will import the libraries we will use
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

### Data splitting

To begin with, we will split the data into two subsets of data to train the model and to test it afterwards and see if the model can generalize correctly the data. We must do this because most of the time we won't have more data in which we can test the model.

```python
X = df_reg['data']
y = df_reg['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

### Calculating the value of $\alpha$

Before we fit the models, we have to take into account that the Lasso and Ridge regressions are sensitive to the value of the hyper-parameter $\alpha$. To do so, we will fit the model with different values of $\alpha$ in a logaritmic scale, so we can clearly see the changes in the precision, using the $R^2$-score.

```python
# First we define a set of values for Alpha in a logaritmic scale to see the changes
alphas = [0.001, 0.01, 0.1, 1, 10, 100]

ridge_r2 = []
lasso_r2 = []

for a in alphas:
    # Pipeline for Ridge
    ridge_pipe = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=a))])
    ridge_pipe.fit(X_train, y_train)
    ridge_r2.append(ridge_pipe.score(X_test, y_test))
    
    # Pipeline for Lasso
    lasso_pipe = Pipeline([('scaler', StandardScaler()), ('model', Lasso(alpha=a))])
    lasso_pipe.fit(X_train, y_train)
    lasso_r2.append(lasso_pipe.score(X_test, y_test))

#Note: We scale the data so the difference in the magnitude of the data doesn't affect the Lasso or the ridge regularization

# Then we present the behavior of the R2 metric according to the different values of alpha
plt.figure(figsize=(10, 6))
plt.plot(alphas, ridge_r2, marker='o', label='Ridge (R²)', color='blue')
plt.plot(alphas, lasso_r2, marker='s', label='Lasso (R²)', color='red')

plt.xscale('log') # Escala logarítmica para mejor visualización
plt.xlabel('Alpha value (logarithmic scale)')
plt.ylabel('R² Score (Precision)')
plt.title('Alpha\'s Effect in the Model Precision')
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.show()
```

As we can see in the previous graph, the ridge regularization tends to be more stable because the coefficients do not reach zero. However, Lasso tends to lose accuracy faster as Alpha increases because it starts to "turn off" variables. This means that the best Alpha is the one where the coefficient is at its maximum in the test set.

In order to know the best value of alpha for both cases, we will use the tool `GridSearchCV` which is a class from the scikit-learn library used for hyperparameter tuning. It systematically looks out for the optimal combination of hyperparameter values given a machine learning model and a list of possible values for each hyperparameter by using cross-validation to evaluate the performance of each combination.

```python
# First we define the possible values of alpha or the grid for the class
param_grid = {'model__alpha': np.logspace(-4, 4, 20)}

# Then we configure the search for the Ridge Regression 
ridge_search = GridSearchCV(
    Pipeline([('scaler', StandardScaler()), ('model', Ridge())]),
    param_grid, cv=5, scoring='r2'
)

# And the Lasso regression
lasso_search = GridSearchCV(
    Pipeline([('scaler', StandardScaler()), ('model', Lasso(max_iter=10000))]),
    param_grid, cv=5, scoring='r2'
)

# Note: In both cases we are using a 5-fold Cross-Validation and we are using the R2 metric to
#  evaluate the performance of the cross-validated model on the test set

# Now we perform the shearch
ridge_search.fit(X_train, y_train)
lasso_search.fit(X_train, y_train)

# Results
print(f"Best alpha for Ridge: {ridge_search.best_params_['model__alpha']:.4f}")
print(f"Best R2 in validation (Ridge): {ridge_search.best_score_:.4f}")
print("-" * 30)
print(f"Best alpha for Ridge: {lasso_search.best_params_['model__alpha']:.4f}")
print(f"Best R2 in validation (Lasso): {lasso_search.best_score_:.4f}")
```

### Comparing the models

Now that we know the best values for $\alpha$ we will use them to compare the metrics of each model.

It's important to note that, before we train the models we use the `StandardScaler` to reduce the impact of the difference in the magnitude of the different units each variable is meassured even though they have almost the same unit, except for age, weight, height, and density. If we do not do this, the Lasso and Ridge regressions will unfairly penalize variables with large numbers, trying to reduce them to zero, while ignoring variables on small scales, even though the latter are the ones that actually predict better.

Also, with this standarization, we can compare which variable is more important by looking at its coefficients, since these won't depend on the unit of measurement.

```python
def fit_reg_models(models_to_fit, X_train, y_train, X_test, y_test, num_variables):
    # This function aims to fit the regression models defined and calculate the errors of the model.
    # It also shows the coeficients of each independent variable and the difference between the predicted and the real values.

    results = []

    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10), sharey='row')

    for i, (name, model) in enumerate(models_to_fit.items()):
        
        pipe = Pipeline(steps=[
            ('scaler', StandardScaler()),
            ('model', model)
        ])
        pipe.fit(X_train, y_train)
        y_pred = pipe.predict(X_test)
        
        # Calculate the metrics
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        
        results.append([name, mse, mae, rmse, r2])
        
        # Now we want to graph the importance of each coefficient and the difference between the
        # model predictions and the test values

        # First we get the column names and each component
        feature_names = X_train.columns
        model_coefficients = pipe.named_steps['model'].coef_

        # Subplot 1: Coefficients
        coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': model_coefficients})
        coef_df = coef_df.sort_values(by='Coefficient', ascending=False)
        
        sns.barplot(x='Coefficient', y='Feature', data=coef_df, hue='Feature', palette='viridis', legend=False, ax=axes[0, i])
        axes[0, i].set_title(f'Importance of the Coefficients ({name} Model)')
        axes[0, i].grid(axis='x', linestyle='--', alpha=0.7)
        if i > 0: axes[0, i].set_ylabel("")

        # Subplot 2: Real values VS. Predicted
        sns.scatterplot(x=y_test, y=y_pred, color='blue', edgecolors='k', ax=axes[1, i])
        
        # Perfect regression line (where prediction == real)
        axes[1, i].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
        axes[1, i].set_xlabel('Real Values (BodyFat %)')
        if i == 0: axes[1, i].set_ylabel("Predicted value (BodyFat %)")
        axes[1, i].set_title(f'Model {name} Fit: Actual vs Predicted')
    
    plt.tight_layout()
    
    plt.show()

    return results
```

To test the performance of the models, we will use the following metrics:
- Root Mean Squared Error (RMSE): It measures the average distance between the predicted and actual values. It penalizes larger errors.
- Mean Absolute Error (MAE): It calculates the average of the absolute differences between predicted and actual values, making it robust to outliers.
- Mean Squared Error (MSE): It squares the errors before averaging, which magnifies the impact of larger errors and helps in understanding the spread of errors.
- R-squared ($R^2$) Score: The $R^2$ score measures the proportion of the variance in the dependent variable (target) that is predictable from the independent variables (features). It provides an indication of how well the model fits the data.

```python
models = {"Linear Regression": LinearRegression(),
          "Lasso": Lasso(alpha=0.0886),
          "Ridge": Ridge(alpha=1.6238)
          }


results = fit_reg_models(models, X_train, y_train, X_test, y_test, "all")

# We create the comparative table
df_results = pd.DataFrame(results, columns=['Model', 'MSE', 'MAE', 'RMSE', 'R2'])
print(df_results.to_markdown(index=False))
```

As we can see, all three models are able to explain the behivor of the body fat ussing all the meassured varaibles, we can say it looking at the $R^2$ score of each model, which is near to 1, meaning that the model us abel to predict almost perfectly the outcome. We can also see this by looking at the scatter graphs that shows how well the model is predicting the outcomes. We identify it by looking at the red dotted line that represents a perfect prediction and the distribution of the points arround it.

The above is also represented by the MSE which is suggesting that the predictions are very close to the actual values. Also the MAE is telling us that, on average, the model is off by only 0.50 percentage points of body fat. Finally, understanding that the RMSE represents the standard deviation of the residuals and the results are similar to the MAE, is pointing out that there are no extremely large errors (outliers) that drastically deviate the model.

Otherwise, we can see that there are coefficients that have more importance to the model, such as Abdomen circumference, Weight and age. But there is an inverse relationship with the density, which we could expect based on the formula used to estimate the body fat:

$$D = \frac{1}{[\left(\frac{A}{a}\right) + \left(\frac{B}{b}\right)]}$$

Where:<br>
D = Body Density (gm/cm^3)<br>
A = proportion of lean body tissue<br>
B = proportion of fat tissue (A+B=1)<br>
a = density of lean body tissue (gm/cm^3)<br>
b = density of fat tissue (gm/cm^3)

## Models used for classification

Now, we will use the diabetes dataset to build some classification models. In this case we will use some binary classification models such as logistic regression, decision tree and perceptron. With these models we want to build a prediction where 1 means that the patient is tested as positive for diabetes.

To see if the model is able to capture and generalize correctly the patients with diabetes, we will use some metrics such as precision, accuarancy, recall, and the f1-score. Aditionally, we will present the confusion matrix in order to show the errors made by the model to assign a label to some data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

### Data splitting

Similarly to the regression model, we will split the data into a training set and into a testing set. Also, as we will use some binary models, we need to change the values of the target variable, so the results are between 0 and 1.

```python
X_class = df_class['data']
y_class = df_class['target']

y_class = y_class.map({'tested_positive': 1, 'tested_negative': 0})

X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, test_size=0.25, random_state=42)
```

### Comparing the models

Now we will fit three different models for classification in order to stimate the conditions of the people diagnosticated with diabetes. To calculate the performance of each model, we will use the precision, recall, accuaracy, and f1-score, understanding each metric as:

- Accuaracy: It's the proportion of all classifications that were correct, whether positive or negative.
- Recall: Also known as the true positive rate (TPR), represents the proportion of all actual positives that were classified correctly as positives.
- Precision: It's the proportion of all the model's positive classifications that are actually positive

To better understand the definitions from above we can use a *Confusion Matrix* which summarizes the performance of a classifier in four essential terms:

- True Positives (TP): Correctly predicted positive instances.
- False Positives (FP): Incorrectly predicted positive instances.
- True Negatives (TN): Correctly predicted negative instances.
- False Negatives (FN): Incorrectly predicted negative instances.

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(max_depth=5, random_state=42),
    "Perceptron": Perceptron(max_iter=1000, random_state=42)
}

fig_cm, axes_cm, = plt.subplots(1, 3, figsize=(20, 5))
fig_pr, axes_pr = plt.subplots(1, 3, figsize=(20, 5))
plt.figure(figsize=(8, 6))
results = []

for i, (name, model) in enumerate(models.items()):
    
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    # Calculate the metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    results.append([name, acc, prec, rec, f1])
    
    # In adition, we calculate the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Then we display it in a graphic
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Tested Negative', 'Tested Positive'])
    disp.plot(ax=axes_cm[i], cmap='Blues', colorbar=False)
    axes_cm[i].set_title(f'Matrix: {name}')
    axes_cm[i].set_xlabel('Prediction')
    axes_cm[i].set_ylabel('Real' if i == 0 else "")

    # Now we will graph the Precision-Recall curve and the ROC curve
    if hasattr(pipe, "predict_proba"):
        y_scores = pipe.predict_proba(X_test)[:, 1]
    else:
        y_scores = pipe.decision_function(X_test)

    # First the Precision-Recall curve
    # We calculate the precision and the recall with diferent thresholds
    precisions, recalls, thresholds = precision_recall_curve(y_test, y_scores)
    avg_precision = average_precision_score(y_test, y_scores)

    # Then we graph the results
    axes_pr[i].plot(recalls, precisions, color='teal', lw=2, label=f'AP (Average Precision) = {avg_precision:.2f}')

    axes_pr[i].fill_between(recalls, precisions, alpha=0.2, color='teal')
    axes_pr[i].set_xlabel('Recall (Capacity to detect diabetes)')
    axes_pr[i].set_ylabel('Precision')
    axes_pr[i].set_title(f'Precision-Recall Curve for: {name}')
    axes_pr[i].grid(True, linestyle='--', alpha=0.6)

    # We draw a line at the target Recall
    axes_pr[i].axvline(x=0.8, color='red', linestyle=':', label='Target Recall (80%)')
    axes_pr[i].legend()

    # Now the ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_scores)
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Referencia: Clasificador aleatorio
plt.plot([0, 1], [0, 1], 'k--', label='Azar (AUC = 0.50)')
plt.xlabel('False Positive Ratio (FPR)')
plt.ylabel('True Positive Ratio (TPR)')
plt.title('Comparison of ROC Curves')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

df_results = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-Score'])
print(df_results.to_markdown(index=False))
```

With the results obtained above we can see that the best model to classificate the data set it's the logistc regression, even though the other models had a good performance. We can say that by looking at the ROC curves where this model is able to find patterns through the data and reduce the amount of false patients diagnosticated with diabetes.

On the other hand, we can see that the Perceptron model were the one with higher number of false negatives, something we won't want in a model of this importance where we would be missing an important and maybe preventive diagnosis. In the Precision-Recall curves we can see that when the models start to detect all patients (Recall close to 1.0), the precision drops rapidly. # Exercise 3: Unsupervised learning.

```python
# First we will call the necesary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
def clean_data(data):
    # Cleaning data
    # Removing the empty values
    data.replace(0, np.nan, inplace=True)
    data.dropna(inplace=True)

    # Removing the Outliers
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    data_filtered = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]

    return data_filtered
```

```python
def get_metrics(data, k_values, variables):
    # With this function we will get the inertia to calculate the number of clustes using the elbow methos
    # and will be calculating the silhoutte score.
    inertia = []
    silhouettes = []

    # We calculate multiple clustes changing the number of k and calculating the inertia and silhoutte score
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(data)
        
        inertia.append(kmeans.inertia_)
        silhouettes.append(silhouette_score(data, kmeans.labels_))

    # Now we graph the results of each metric compared to the number of clusters specified
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    ax[0].plot(range(min(k_values), max(k_values)+1), inertia, 'o-', label=f'With {variables} Variables')
    ax[0].set_title('Elbow method (Inertia)')
    ax[0].set_xlabel('Number of clusters (k)')
    ax[0].set_ylabel('Inertia')
    ax[0].legend()

    ax[1].plot(range(min(k_values), max(k_values)+1), silhouettes, 'o-', color='green', label=f'With {variables} Variables')
    ax[1].set_xlabel('Number of clusters (k)')
    ax[1].set_ylabel('Quality')
    ax[1].set_title('Silhouette Score')
    ax[1].legend()

    plt.show()
    
    return inertia, silhouettes

def kmeans_cluster(data, k_value, column_names, title):
    # With this function we will calculate the clustering of the data introduced and define some
    # characteristics of the clusters such as the centroids and the number of points of each cluster

    # We scale the data to avoid problems with different scales of meassurements
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(data)

    # We create a copy of the dataset so we do not rewrite it by mistake
    data = data.copy()

    # We create the clusters by the k defined
    kmeans = KMeans(n_clusters=k_value, random_state=42, n_init=10)
    kmeans.fit(X_scaled)

    # We add the labels created for each cluster
    data.loc[:, 'Cluster'] = kmeans.labels_

    # We graph the results
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=column_names[0], y=column_names[1], hue='Cluster', palette='viridis', s=100)
    plt.scatter(scaler.inverse_transform(kmeans.cluster_centers_)[:, 0], 
                scaler.inverse_transform(kmeans.cluster_centers_)[:, 1], 
                s=300, c='red', marker='X', label='Centroids')

    plt.title(title)
    plt.legend()
    plt.show()

    # Now we present the centroids of each cluster and the amount of data of each cluster
    centroids = scaler.inverse_transform(kmeans.cluster_centers_)
    data_byClusters = data['Cluster'].value_counts().sort_index()

    print(f"{'Cluster':<10} | {column_names[0]:<10} | {column_names[1]:<10} | {'Amount of Patients'}")
    print("-" * 65)

    for i in range(len(centroids)):
        glucose = centroids[i][0]
        bmi = centroids[i][1]
        amount = data_byClusters[i]
        
        print(f"{i:<10} | {glucose:<10.2f} | {bmi:<10.2f} | {amount} pacientes")
```

### Selecting the best k-number

For this example we will use the elbow method to estimate the best number of clusters by looking for a point where the improvement of the WCSS (Within-Cluster Sum of Squares), which shows how close the data points are to their cluster centroids, slows down even if we increase the number of clusters. For this case, we will use the inertia which represents the sum of squared distances of each data point to its closest cluster center.

Also, we will look at the Silhouette Score given that it's a metric used to evaluate the quality of clustering results. It measures how similar each data point is to its own cluster compared to other clusters, helping assess how well the data has been grouped. In this case values near to 1 inidcates a well clustered data, around 0 it means that is near cluster boundaries or clusters overlap. And, close to -1 it means that the cluster is poor.

```python
# We define a range of k-numbers to look out for the best result
k_range=range(2, 10)

X_class_scaled = StandardScaler().fit_transform(X_class_kmeans)

inertia, silhouettes = get_metrics(X_class_scaled,  k_range, 2)
```

With the previous graphs we can see that the model has problems to correctly cluster the data. For the Silhouette Score we can see that it cannot reach values above 0.4, which indicates that the model has problems to create a corret cluster. This could explains why the elbow method cannot reach an optimal point even with 5 or more clusters, if we keep increasing the number of cluster we will end up overfittig our model. In this case, we will try to use 3 clusters because it has the highest Silhouette Score.

```python
# Now we graph the results
kmeans_cluster(X_class_kmeans, 3, ['Glucose', 'BMI'], 'Clustering patiens with: Glucose vs. BMI')
```

As we could expected, the model has some problems to cluster the data because the groups made don't really represent the reallity of the diabetes problem. If we look at the cluster number 2, with 181 patients, we can say that it represents the people with diabetes, because the have a high level of glucosa. But yet at a level of pre-diabetes.

However, the groups 0, with 257 patients, and 1, with 306 patients, has some problems to stablish a clear frontier of patients. Both clusters take people with healthy and unhealthy parameters which hinders the classification of each group.

We can also see that each cluster has a similiar amount of patients which can tell us that the model is equally distributed and each cluster is representative.
