# Exercise 3: Unsupervised learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python

data = fetch_openml(data_id=37, as_frame=True, parser='auto').frame
# Seleccionamos las dos variables más relevantes
df_clustering = data[['plas', 'mass']].copy() # plas=Glucosa, mass=BMI
df_clustering.columns = ['Glucose', 'BMI']

# Reemplazar ceros por NaN y eliminar filas incompletas
df_clustering.replace(0, np.nan, inplace=True)
df_clustering.dropna(inplace=True)

# Eliminar Outliers (Método IQR)
Q1 = df_clustering.quantile(0.25)
Q3 = df_clustering.quantile(0.75)
IQR = Q3 - Q1
df_filtered = df_clustering[~((df_clustering < (Q1 - 1.5 * IQR)) | (df_clustering > (Q3 + 1.5 * IQR))).any(axis=1)]

# Estandarización (Crucial para K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_filtered)
```

```python
inercia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inercia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K_range, inercia, 'bx-')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para determinar k óptimo')
plt.show()
```

```python
# Aplicar K-means con k=3
kmeans_final = KMeans(n_clusters=3, random_state=42, n_init=10)
df_filtered['Cluster'] = kmeans_final.fit_transform(X_scaled).argmax(axis=1) # Asignar etiquetas

# Graficar
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='Glucose', y='BMI', hue='Cluster', palette='viridis', s=100)
plt.scatter(scaler.inverse_transform(kmeans_final.cluster_centers_)[:, 0], 
            scaler.inverse_transform(kmeans_final.cluster_centers_)[:, 1], 
            s=300, c='red', marker='X', label='Centroides')

plt.title('Segmentación de Pacientes: Glucosa vs BMI')
plt.legend()
plt.show()

# Establecer Perfiles (Promedios por cluster)
print(df_filtered.groupby('Cluster').mean())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1. Preparar datos (Limpieza similar al paso anterior)
df_all = data.replace(0, np.nan).dropna()
X_all_scaled = StandardScaler().fit_transform(df_all.drop(columns=['class']))
X_2vars_scaled = X_scaled # El de Glucosa e IMC que ya tenemos

# Función para calcular métricas
def get_metrics(data, k_range=range(2, 6)):
    inertia = []
    silhouettes = []
    for k in k_range:
        km = KMeans(n_clusters=k, random_state=42, n_init=10).fit(data)
        inertia.append(km.inertia_)
        silhouettes.append(silhouette_score(data, km.labels_))
    return inertia, silhouettes

in_2, sil_2 = get_metrics(X_2vars_scaled)
in_all, sil_all = get_metrics(X_all_scaled)

# Visualización
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico de Inercia (Codo)
ax[0].plot(range(2, 6), in_2, 'o-', label='2 Variables')
ax[0].plot(range(2, 6), in_all, 's-', label='Todas las Variables')
ax[0].set_title('Método del Codo (Inercia)')
ax[0].legend()

# Gráfico de Coeficiente de Silueta
ax[1].plot(range(2, 6), sil_2, 'o-', color='green', label='2 Variables')
ax[1].plot(range(2, 6), sil_all, 's-', color='red', label='Todas las Variables')
ax[1].set_title('Coeficiente de Silueta (Calidad)')
ax[1].legend()

plt.show()
```
